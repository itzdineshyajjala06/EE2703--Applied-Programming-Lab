## Polygon Morphing

```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'r')

def init():
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    return ln,

def update(frame):
    if frame <= 0.5:
        alpha = frame * 2
        if alpha <= 0.2:
            xdata, ydata = morph(xs, ys, xt, yt, alpha*5)
        elif alpha <= 0.4:
            xdata, ydata = morph(xp, yp, xs, ys, (alpha-0.2)*5)
        elif alpha <= 0.6:
            xdata, ydata = morph(xh, yh, xp, yp, (alpha-0.4)*5)
        elif alpha <=0.8:
            xdata, ydata = morph(xhp, yhp, xh, yh, (alpha-0.6)*5)
        else:
            xdata, ydata = morph(xo, yo, xhp, yhp, (alpha-0.8)*5)
    else:
        alpha = (frame - 0.5) * 2
        if alpha <=0.2:
            xdata, ydata = morph(xhp, yhp, xo, yo, (alpha)*5)
        elif alpha <=0.4:
            xdata, ydata = morph(xh, yh, xhp, yhp, (alpha-0.2)*5)
        elif alpha <=0.6:
            xdata, ydata = morph(xp, yp, xh, yh, (alpha-0.4)*5)
        elif alpha <=0.8:
            xdata, ydata = morph(xs, ys, xp, yp, (alpha-0.6)*5)
        else:
            xdata, ydata = morph(xt, yt, xs, ys, (alpha-0.8)*5)
    ln.set_data(xdata, ydata)
    return ln,

def triangle(t):
    n3= round(int(len(t)/3))
    ts = np.linspace(-0.5, 1, n3)
    rs1 = np.linspace(-np.sqrt(3)/2,np.sqrt(3)/2,n3)
    rs2 = np.linspace(np.sqrt(3)/2,0,n3)
    rs3 = np.linspace(-np.sqrt(3)/2,0,n3)
    xt = np.concatenate([ts[::-1],-0.5*np.ones(n3),ts])
    yt = np.concatenate([rs3[::-1],rs1,rs2])
    return xt,yt

def square(t):
    n4 = int(len(t)/4)
    ts1 = np.linspace(-1,0, n4)
    ts2 = np.linspace(0,1, n4)
    xs = np.concatenate([ts2[::-1],ts1[::-1],ts1,ts2])
    ys = np.concatenate([ts1[::-1],ts1,ts2,ts2[::-1]])
    return xs, ys

def pentagon(t):
    n5 = int(len(t)/5)
    a1=2*np.pi/5
    a2=4*np.pi/5
    a3=6*np.pi/5
    a4=8*np.pi/5
    x1=np.linspace(1,np.cos(a4),n5)
    x2=np.linspace(np.cos(a4),np.cos(a3),n5)
    x3=np.linspace(np.cos(a3),np.cos(a2),n5)
    x4=np.linspace(np.cos(a2),np.cos(a1),n5)
    x5=np.linspace(np.cos(a1),1,n5)
    y1=np.linspace(0,np.sin(a4),n5)
    y2=np.linspace(np.sin(a4),np.sin(a3),n5)
    y3=np.linspace(np.sin(a3),np.sin(a2),n5)
    y4=np.linspace(np.sin(a2),np.sin(a1),n5)
    y5=np.linspace(np.sin(a1),0,n5)
    xp = np.concatenate([x1,x2,x3,x4,x5])
    yp = np.concatenate([y1,y2,y3,y4,y5])
    return xp,yp

def hexagon(t):
    n6 = int(len(t)/6)
    a1=2*np.pi/6
    a2=4*np.pi/6
    a3=6*np.pi/6
    a4=8*np.pi/6
    a5=10*np.pi/6
    x1=np.linspace(1,np.cos(a5),n6)
    x2=np.linspace(np.cos(a5),np.cos(a4),n6)
    x3=np.linspace(np.cos(a4),np.cos(a3),n6)
    x4=np.linspace(np.cos(a3),np.cos(a2),n6)
    x5=np.linspace(np.cos(a2),np.cos(a1),n6)
    x6=np.linspace(np.cos(a1),1,n6)
    y1=np.linspace(0,np.sin(a5),n6)
    y2=np.linspace(np.sin(a5),np.sin(a4),n6)
    y3=np.linspace(np.sin(a4),np.sin(a3),n6)
    y4=np.linspace(np.sin(a3),np.sin(a2),n6)
    y5=np.linspace(np.sin(a2),np.sin(a1),n6)
    y6=np.linspace(np.sin(a1),0,n6)
    xh = np.concatenate([x1,x2,x3,x4,x5,x6])
    yh = np.concatenate([y1,y2,y3,y4,y5,y6])
    return xh,yh

def heptagon(t):
    n7 = int(len(t)/7)
    a1=2*np.pi/7
    a2=4*np.pi/7
    a3=6*np.pi/7
    a4=8*np.pi/7
    a5=10*np.pi/7
    a6=12*np.pi/7
    x1=np.linspace(1,np.cos(a6),n7)
    x2=np.linspace(np.cos(a6),np.cos(a5),n7)
    x3=np.linspace(np.cos(a5),np.cos(a4),n7)
    x4=np.linspace(np.cos(a4),np.cos(a3),n7)
    x5=np.linspace(np.cos(a3),np.cos(a2),n7)
    x6=np.linspace(np.cos(a2),np.cos(a1),n7)
    x7=np.linspace(np.cos(a1),1,n7)
    y1=np.linspace(0,np.sin(a6),n7)
    y2=np.linspace(np.sin(a6),np.sin(a5),n7)
    y3=np.linspace(np.sin(a5),np.sin(a4),n7)
    y4=np.linspace(np.sin(a4),np.sin(a3),n7)
    y5=np.linspace(np.sin(a3),np.sin(a2),n7)
    y6=np.linspace(np.sin(a2),np.sin(a1),n7)
    y7=np.linspace(np.sin(a1),0,n7)
    xhp = np.concatenate([x1,x2,x3,x4,x5,x6,x7])
    yhp = np.concatenate([y1,y2,y3,y4,y5,y6,y7])
    return xhp,yhp

def octagon(t):
    n8 = int(len(t)/8)
    a1=2*np.pi/8
    a2=4*np.pi/8
    a3=6*np.pi/8
    a4=8*np.pi/8
    a5=10*np.pi/8
    a6=12*np.pi/8
    a7=14*np.pi/8
    x1=np.linspace(1,np.cos(a7),n8)
    x2=np.linspace(np.cos(a7),np.cos(a6),n8)
    x3=np.linspace(np.cos(a6),np.cos(a5),n8)
    x4=np.linspace(np.cos(a5),np.cos(a4),n8)
    x5=np.linspace(np.cos(a4),np.cos(a3),n8)
    x6=np.linspace(np.cos(a3),np.cos(a2),n8)
    x7=np.linspace(np.cos(a2),np.cos(a1),n8)
    x8=np.linspace(np.cos(a1),1,n8)
    y1=np.linspace(0,np.sin(a7),n8)
    y2=np.linspace(np.sin(a7),np.sin(a6),n8)
    y3=np.linspace(np.sin(a6),np.sin(a5),n8)
    y4=np.linspace(np.sin(a5),np.sin(a4),n8)
    y5=np.linspace(np.sin(a4),np.sin(a3),n8)
    y6=np.linspace(np.sin(a3),np.sin(a2),n8)
    y7=np.linspace(np.sin(a2),np.sin(a1),n8)
    y8=np.linspace(np.sin(a1),0,n8)
    xo = np.concatenate([x1,x2,x3,x4,x5,x6,x7,x8])
    yo = np.concatenate([y1,y2,y3,y4,y5,y6,y7,y8])
    return xo,yo

def morph(x1, y1, x2, y2, alpha):
    xm = alpha * x1 + (1-alpha) * x2
    ym = alpha * y1 + (1-alpha) * y2
    return xm, ym

t = np.linspace(-np.sqrt(3)/2, np.sqrt(3)/2, 840)

xt, yt = triangle(t)
xs, ys = square(t)
xp,yp = pentagon(t)
xh,yh = hexagon(t)
xhp,yhp = heptagon(t)
xo,yo = octagon(t)

ani = FuncAnimation(fig, update, frames=np.linspace(0, 1, 800),init_func=init, blit=True, interval=10, repeat=False)
plt.show()
```

## Explaination
By observing the video (*Given in the Moodle*), I understand that the vertice **(1,0)** is common for all the polygons. So, keeping that in the mind, I started the side drawings from that vertice for every polygon. I defined all the required polygons in their respective functions and used **update** function to update the frame every 10 miliseconds.Since, it is required to remorph the polygons in reverse order I divided the frames accordingly like for **frame 0 - 0.5** it morphs from triangle to square, then to pentagon next hexagon then heptagon and last octagon and for **frame 0.5-1** I reversed the order. I make sure that **alpha** in **morph** function will range from **0-1** by rearranging it. We can adjust the speed of this morphing by chaning the **frames** number in the **FunAnimation** command. I stopped the repeation of this morphing by giving **False** in the **repeat** command inside FunAnimation. The scaling of the graph was limited to the range **[-1.2,1.2]** for both x-axis and y-axis by defining it in the **init** function. Finally, the significance of **len(t)** is very important. So, I took the least multiple of [3,4,5,6,7,8] (*the number of sides of the polygons used*) which is **840**. Always, the *len(t)* should be a multiple of 840.
